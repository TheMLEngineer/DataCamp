# Heads and Tails Toss Simulation

```python
np.random.seed(123)
final_heads = []
for i in range(100):
    heads = [0]
    
    for i in range(10):
        coin = np.random.randint( 0 , 2 )
        heads.append(heads[i] + coin)
    final_heads.append(heads[-1])
```

```python
print(final_heads)
```

```python
# So heads are thrown 7 times often
```

```python
plt.hist(final_heads)
plt.show()
```

Curve becomes more smooth when we run for more times

```python
np.random.seed(123)
final_heads = []
for i in range(100000):
    heads = [0]
    
    for i in range(10):
        coin = np.random.randint( 0 , 2 )
        heads.append(heads[i] + coin)
    final_heads.append(heads[-1])
```

```python
print(final_heads)
```

```python
# So heads are thrown 4 times often
```

```python
plt.hist(final_heads)
plt.show()
```

3 Scopes 

1) Builtin scope (print fn , sum fn etc)

2) Global scope (All program)

3) Local Scope (Specific to fn)


First program looks into Local scope , then global , then built in
