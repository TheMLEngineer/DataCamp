**Pickling**

It will convert OBJ --> Byte stream (For dictionary pickling will create JSON)

it serializes objects so they can be saved to a file, and loaded in a program again later on.

```python
import pickle
dict1 = {1 : 'hi' , 
        2 : 'hello'}
filename = 'pickle'
outfile = open(filename , 'wb')
#Write binary
pickle.dump(dict1 , outfile)
outfile.close()
with open('pickle.pkl' , 'rb') as f:#read binary
    data = pickle.load(f)
print(data)
```

**.pkl file is not human readable , so provides some security too**
